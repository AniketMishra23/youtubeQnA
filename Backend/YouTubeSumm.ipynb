```python
# Importing all the required libraries
from youtube_transcript_api import YouTubeTranscriptApi

# Using BART model from facebook
from transformers import pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# Creating a function to extract transcript from a video
def get_video_transcript(id):
  transcript = YouTubeTranscriptApi.get_transcript(id)
  script = ""
  word_count = 0 # Starting a counter from 0

  for text in transcript:
    t = text["text"]
    if t != '[Music]':
      # Split the text into words and add them to the script
      words = t.split()
      for word in words:
        script += word + " "
        word_count += 1
        # Check and break the loop if word limit is reached
        if word_count >= 600:
          break

  return script
```

### Sample link :- https://www.youtube.com/watch?v=1oj3uLVTH4o

```python
video_url = input("Enter the YouTube video URL: ")
id = video_url.split("=")[1] # Extracting video ID from URL
# print(id)
```

```python
# Calling get_video_transcript function for transcript
trans = get_video_transcript(id)
# print(trans)
```

```python
# Embedding the full transcript in BART model to provide a short summary of the YouTube video
summary = summarizer(trans, max_length=130, min_length=30, do_sample=False)
print(summary[0]['summary_text'])
```
